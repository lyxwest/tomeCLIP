```python
import matplotlib.pyplot as plt
import numpy as np
```

### k400 zero-shot on HMDB

```python
ratio = ['baseline', '4', '8', '12', '16']

acc = np.array((52.810, 52.157, 52.418, 36.471, 3.464))
speed = 1 / np.array((0.2863, 0.2350, 0.1800, 0.1299, 0.0990))
speed = np.around(speed, 3)
```

```python

fig,ax=plt.subplots(1,2, figsize=(8, 4))

ax[0].set_title('ratio-acc')
ax[0].set_xlabel('ratio')
ax[0].set_ylabel('acc(%)')


# 设置数字标签
for a, b in zip(ratio, acc):
    ax[0].text(a, b, b, fontsize=10)

ax[0].plot(ratio, acc)

# 设置数字标签
for a, b in zip(ratio, speed):
    ax[1].text(a, b, b, fontsize=10)
ax[1].set_title('ratio-speed')
ax[1].set_xlabel('ratio')
ax[1].set_ylabel('speed(vid/s)')

ax[1].plot(ratio, speed)
```

### HMDB VIT base

```python
ratio = ['baseline', '4', '8', '12', '16']

acc_1gpu = np.array((72.288, 72.092, 70.850, 71.438, 68.235))
speed_1gpu = 1 / np.array((0.0693, 0.0640, 0.0558, 0.0476, 0.0395))
speed_up_1gpu = (speed_1gpu[1:] - speed_1gpu[0]) / speed_1gpu[0]
speed_1gpu = np.around(speed_1gpu, 3)
speed_up_1gpu = np.around(speed_up_1gpu, 3) * 100

acc_4gpu = np.array((72.258, 72.063, 70.822, 71.410, 68.211))
speed_4gpu = 1 / np.array((0.0211, 0.0204, 0.0184, 0.0169, 0.0157))
speed_up_4gpu = (speed_4gpu[1:] - speed_4gpu[0]) / speed_4gpu[0]
speed_4gpu = np.around(speed_4gpu, 3)
speed_up_4gpu = np.around(speed_up_4gpu, 3) * 100
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12,9))

ax[0][0].set_title('ratio-acc(1 gpu)')
ax[0][0].set_xlabel('ratio')
ax[0][0].set_ylabel('acc(%)')


# 设置数字标签
for a, b in zip(ratio, acc_1gpu):
    ax[0][0].text(a, b, b, fontsize=10)

ax[0][0].plot(ratio, acc_1gpu)

# 设置数字标签
a_l = -1
b_l = -1
count = 0

for a, b in zip(ratio, speed_1gpu):
    ax[0][1].text(a, b, b, fontsize=10)
    if a_l != -1 and b_l != -1:
        ax[0][1].text(a, b - 0.5, '({}%)'.format(speed_up_1gpu[count]), fontsize=10)
        count += 1
        a_l = a
        b_l = b      
    else:
        a_l = a
        b_l = b
    
    
    

ax[0][1].set_title('ratio-speed(1 gpu)')
ax[0][1].set_xlabel('ratio')
ax[0][1].set_ylabel('speed(vid/s)')

ax[0][1].plot(ratio, speed_1gpu)


# 设置数字标签
for a, b in zip(ratio, acc_4gpu):
    ax[1][0].text(a, b, b, fontsize=10)
    

ax[1][0].set_title('ratio-acc(4 gpu)')
ax[1][0].set_xlabel('ratio')
ax[1][0].set_ylabel('speed(vid/s)')

ax[1][0].plot(ratio, acc_4gpu)



# 设置数字标签
a_l = -1
b_l = -1
count = 0

for a, b in zip(ratio, speed_4gpu):
    ax[1][1].text(a, b, b, fontsize=10)
    if a_l != -1 and b_l != -1:
        ax[1][1].text(a, b - 1, '({}%)'.format(speed_up_4gpu[count]), fontsize=10)
        count += 1
        a_l = a
        b_l = b      
    else:
        a_l = a
        b_l = b

ax[1][1].set_title('ratio-speed(4 gpu)')
ax[1][1].set_xlabel('ratio')
ax[1][1].set_ylabel('speed(vid/s)')

ax[1][1].plot(ratio, speed_4gpu)
```
